**Student Mark Analyzer**

```python
import numpy as np
stu=int(input("Enter the no of Students\t:"))
sub=int(input("Enter the no of Subjects\t:"))
marks=np.zeros((stu,sub))
for i in range(stu):
  for j in range(sub):
    marks[i][j]=int(input(f"Enter the marks of student {i+1} in subject {j+1} :"))
print(marks)
```

```python
# The avg marks of the each student

avg_mark_stu=np.mean(marks,axis=1)
print(np.round(avg_mark_stu,2))

# The avg mark of each subject
avg_marks_sub=np.mean(marks,axis=0)
print(np.round(avg_marks_sub,2))
```

```python
# The total mark of each student
total_mark_stu=np.sum(marks,axis=1)
print(total_mark_stu)
```

```python
# The maximum mark of each student
max_mark_stu=np.max(marks,axis=1)
print(max_mark_stu)
# The highest mark in each subject
max_mark_sub=np.max(marks,axis=0)
print(max_mark_sub)
```

```python
# standard deviation of each subject
std_dev_sub=np.std(marks,axis=0)
print(std_dev_sub)
std_dev_stu=np.std(marks,axis=1)
print(std_dev_stu)
```

```python
# adding new students mark to the array
c=[65,67,69]
marks=np.vstack((marks,c))
print(marks)
```

```python
# max-min normalizarion
norm=((marks-np.min(marks))/(np.max(marks)-np.min(marks)))
np.round(norm,2)
```

```python
# z score normalization
Z_norm=((marks-np.mean(marks))/np.std(marks))
np.round(Z_norm,2)
```
