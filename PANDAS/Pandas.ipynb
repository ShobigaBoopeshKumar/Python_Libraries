```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Step 1: Create your own dataset (10 students, 3 subjects)
X = np.array([
    [85, 70, 88],   # Student 1
    [78, 65, 80],   # Student 2
    [92, 88, 95],   # Student 3
    [70, 60, 72],   # Student 4
    [88, 85, 90],   # Student 5
    [65, 55, 60],   # Student 6
    [90, 82, 94],   # Student 7
    [72, 68, 75],   # Student 8
    [80, 70, 85],   # Student 9
    [95, 90, 98]    # Student 10
])

# Step 2: Apply PCA (reduce from 3D  2D)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Step 3: Plot the transformed data
plt.figure(figsize=(8,6))
plt.scatter(X_pca[:,0], X_pca[:,1], color='blue')

for i in range(len(X)):
    plt.text(X_pca[i,0]+0.1, X_pca[i,1]+0.1, f"Student {i+1}")

plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA on Custom Student Scores Dataset")
plt.show()

# Step 4: Print explained variance
print("Explained variance ratio:", pca.explained_variance_ratio_)
```
